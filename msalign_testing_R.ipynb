```r
library(ggplot2)
library(gridExtra)
library(dplyr)
library(reshape2)
```

```r
data <- read.csv(file = 'align-test-output.csv', header = F)
```

```r
data
```

```r
t_data = t(data)
colnames(t_data) <- paste0("COLUMN", seq(10))
rownames(t_data) <- seq(100)
t_data = data.frame(t_data)
```

```r
t_data = t_data %>% mutate(Xax = seq(100))
```

```r
t_data
```

```r
p_1 <- ggplot(t_data, aes(x = seq(100), y = COLUMN1)) + geom_line() + theme(axis.text.x = element_blank(),
        axis.ticks = element_blank(),
        axis.title.x = element_blank())

p_2 <- ggplot(t_data, aes(x = seq(100), y = COLUMN2 )) + geom_line() + theme(axis.text.x = element_blank(),
        axis.ticks = element_blank(),
        axis.title.x = element_blank())
```

```r
p_1
```

```r
grid.arrange(p_1, p_2, nrow = 2, heights = c(0.35, 0.65))
```
